```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
onlinenews['weekday_is_monday'].plot(kind='hist')
plt.title('What day was the article published?')
plt.xlabel('Day')
plt.show()
```

```python
import seaborn as sns
plot=sns.kdeplot(onlinenews['rate_positive_words'], shade=True, color="r")
plot=sns.kdeplot(onlinenews['rate_negative_words'], shade=True, color="b")
plot.set(title = 'Rate of Positive and Negative Words', xlabel='Rate', ylabel='Density')
```
