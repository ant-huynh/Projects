```python
masked_variable = weekday[ 'value'] == 1
```

```python
online_news_data_subject = onlinenews_data[[' data_channel_is_lifestyle', ' data_channel_is_entertainment', ' data_channel_is_bus', ' data_channel_is_socmed', ' data_channel_is_tech', ' data_channel_is_world']]
```

```python
subject = pd.melt(online_news_data_subject)
```

```python
masked_variable = subject[ 'value'] == 1
```

```python
subject_1=subject.loc[masked_variable, 'variable']
```

```python
subject_1.value_counts().plot(kind='bar')
plt.title('Number of Articles Published Per Subject')
plt.xlabel('Subject')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
```
